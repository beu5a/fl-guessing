```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
```

```python
root_dir = '/mnt/nfs/dhasade/gel_logs/fedavg/tuning/shakespeare'
```

```python
files = os.listdir(root_dir)
```

```python
ta = 'test_accuracies'
tl = 'test_losses'
ti = 'test_index'
tra = 'train_accuracies'
trl = 'train_losses'
tri = 'train_index'
```

```python
x = {}
lrs = []
for file in files:
    if(file.startswith('10')):
        lr = file.split('_')[-1][2:]
        lrs.append(lr)

        test_file_path = os.path.join(root_dir, file, 'r1', 'test.csv')
        df = pd.read_csv(test_file_path)
        x[lr] = {ta: df[ta].to_numpy(), tl: df[tl].to_numpy(), ti: df[ti].to_numpy()}
        
        train_file_path = os.path.join(root_dir, file, 'r1', 'train.csv')
        df = pd.read_csv(train_file_path)
        x[lr].update({tra: df[tra].to_numpy(), trl: df[trl].to_numpy(), tri: df[tri].to_numpy()})
```

```python
fig, ((axs1, axs2), (axs3, axs4)) = plt.subplots(2, 2, figsize=(15, 10), sharex=True)
for lr in lrs:
    axs1.plot(x[lr][ti], x[lr][ta], label=lr)
    axs2.plot(x[lr][ti], x[lr][tl], label=lr)
    axs3.plot(x[lr][tri], x[lr][tra], label=lr)
    axs4.plot(x[lr][tri], x[lr][trl], label=lr)

axs1.legend()
axs1.set_title('Test Accuracy')

axs2.set_title('Test Loss')

axs3.legend()
axs3.set_title('Train Accuracy')

axs4.set_title('Train Loss')

fig.supxlabel('Round Number')
fig.suptitle('FedAvg: Hyperparameter Tuning (SGD with momemtum) Shakespeare')
# plt.savefig('../logs/figures/fedavg_shakespeare_hyp_tuning.jpg', dpi=300)
```

```python
# r1 = '/home/dhasade/FL/efficient-federated-learning/logs/fedavg/guessing/femnist/target'
# r2 = '/home/dhasade/FL/efficient-federated-learning/logs/fedavg/guessing/femnist/8_24_gp30'
# r3 = '/home/dhasade/FL/efficient-federated-learning/logs/fedavg/guessing/femnist/8_24_gx30'

# r2 = '/home/dhasade/FL/efficient-federated-learning/logs/fedavg/guessing/synthetic/4_13_gp30_0.01'
# r3= '/home/dhasade/FL/efficient-federated-learning/logs/fedavg/guessing/synthetic/4_13_gx30_0.01'
# r1 = '/home/dhasade/FL/efficient-federated-learning/logs/fedavg/guessing/synthetic/4_13_g0_0.01'

r2 = '/mnt/nfs/dhasade/gel_logs/fedavg/tuning/shakespeare/10_30_gg30_lr0.3'
r3 = '/mnt/nfs/dhasade/gel_logs/fedavg/guessing/shakespeare/10_30_g0_lr0.3'
```

```python
fa = 'fedavg'
g = 'GeL'
fa_b = 'FedAvg Baseline'
alpha = 0.4
```

```python
x = {}
y = {}
y_error = {}

for r, alg in zip([r2, r3], [g, fa_b]):
    tas = []
    tls = []
    tis = []

    tras = []
    trls = []
    tris = []

    for run_num in os.listdir(r):
    
        test_file_path = os.path.join(r, run_num, 'test.csv')
        df = pd.read_csv(test_file_path)
        df = df[df[ti] < 750]
        tas.append(df[ta].to_numpy()); tls.append(df[tl].to_numpy()); tis.append(df[ti].to_numpy())

        train_file_path = os.path.join(r, run_num, 'train.csv')
        df = pd.read_csv(train_file_path)
        df = df[df[tri] < 750]
        tras.append(df[tra].to_numpy()); trls.append(df[trl].to_numpy()); tris.append(df[tri].to_numpy())

    tas = np.array(tas); tls = np.array(tls); tis = np.array(tis)
    tras = np.array(tras); trls = np.array(trls);  tris = np.array(tris)

    y[alg] = {ta: np.mean(tas, axis=0), tl: np.mean(tls, axis=0), tra: np.mean(tras, axis=0), trl: np.mean(trls, axis=0)}
    y_error[alg] = {ta: np.std(tas, axis=0), tl: np.std(tls, axis=0), tra: np.std(tras, axis=0), trl: np.std(trls, axis=0)}
    x[alg] = {ti: tis[0], tri: tris[0]}
```

```python
base_dir = '/home/dhasade/FL/efficient-federated-learning/logs/plotting_csv/'
for alg in [g]:
    all_test = {}
    all_train = {}
    for k,v in y[alg].items():
        if(k == ta or k == tl):
            all_test[k+'_mean'] = v
        else:
            all_train[k+'_mean'] = v
    for k,v in y_error[alg].items():
        if(k == ta or k == tl):
            all_test[k+'_std'] = v
        else:
            all_train[k+'_std'] = v
    for k,v in x[alg].items():
        if(k == ti):
            all_test[k] = v
        else:
            all_train[k] = v
    
    test_filename = alg.replace(" ", "_") + '_test1.csv'
    train_filename = alg.replace(" ", "_") + '_train1.csv'
    df1 = pd.DataFrame(all_test)
    df2 = pd.DataFrame(all_train)
```

## For charting total computation bar plot

```python
r1 = '/home/dhasade/FL/efficient-federated-learning/logs/fedavg/guessing/femnist/target'
r2 = '/home/dhasade/FL/efficient-federated-learning/logs/fedavg/guessing/femnist/8_24_gp30'
r3 = '/home/dhasade/FL/efficient-federated-learning/logs/fedavg/guessing/femnist/8_24_gx30'
```

```python
fa = 'fedavg'
g = 'gel (g = 30%)'
fa_b = 'baseline'
alpha = 0.4
```

```python
y = []
y_error = []

for r, alg in zip([r1, r2, r3], [fa, g, fa_b]):
    bs = []

    for run_num in os.listdir(r):
        budgets_file_path = os.path.join(r, run_num, 'budgets.csv')
        df = pd.read_csv(budgets_file_path, header=None)
        bs.append(df.sum())

    bs = np.array(bs)

    y.append(np.mean(bs))
    y_error.append(np.std(bs))

y = np.array(y)
y_error = np.array(y_error)
```

```python
styles = plt.style.available
```

```python
# plt.style.use(styles[1])
plt.figure(figsize=(6, 7))
plt.bar([fa, g, fa_b], y, yerr = y_error, width=0.8, color=['#1f77b4', '#ff7f0e', '#2ca02c'], alpha=0.7)
plt.ylabel('Total Gradient Computations')
plt.title('Comparing computations FedAvg vs GeL')
# plt.savefig('../logs/figures/computation_femnist_p30.jpg', dpi=300)
```
